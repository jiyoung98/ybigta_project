```
X = []
for i in data.LSTM_pp:
  xinnerlist = i.split(',')
  X.append(xinnerlist)

X
```

```
from gensim.models import Word2Vec
model = Word2Vec(sentences = X, size = 100, window = 5, min_count = 20, workers = 4, sg = 0, iter=10)
# size : 차원 사이즈, min_count: mincount 이하로 등장하는 단어는 무시, sg: 0이면 CBOW방식, 1이면 skipgram방식 활용, worker:실행할 병렬 프로세수 수, iter=epoch, window=주변 단어 개수
```

```
# 유클리디안 거리 구하기!
def euclidian_dist(p1,p2):
  #p1 = embedding_model.wv.vectors[1]
  #p2 = embedding_model.wv.vectors[5]

  dist = np.linalg.norm(p1 - p2) 
  return dist
```

```
outerlist = []
for i, word in enumerate(vect):
  innerlist = []
  for j, compare in enumerate(vect):
    innerlist.append(euclidian_dist(vect[i],vect[j]))
  outerlist.append(innerlist)

dist_matrix = np.array(outerlist)
dist_matrix = pd.DataFrame(dist_matrix)
```
