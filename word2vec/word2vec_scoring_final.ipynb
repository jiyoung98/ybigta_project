```
# 유클리디안 거리 구하기!
def euclidian_dist(p1,p2):
  #p1 = embedding_model.wv.vectors[1]
  #p2 = embedding_model.wv.vectors[5]

  dist = np.linalg.norm(p1 - p2) 
  return dist
```

```
outerlist = []
for i, word in enumerate(vect):
  innerlist = []
  for j, compare in enumerate(vect):
    innerlist.append(euclidian_dist(vect[i],vect[j]))
  outerlist.append(innerlist)

dist_matrix = np.array(outerlist)
dist_matrix = pd.DataFrame(dist_matrix)
```

```
dist_unnamed = dist_matrix
```

```
temp = []
for i in range(len(dist_unnamed)):
  for j in range(len(dist_unnamed)):
    temp.append(dist_unnamed[i][j])

var = np.var(temp)
var

```
