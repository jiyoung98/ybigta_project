```
def make_corpus_for_gensim(untokenized_review_list):
  for sent in untokenized_review_list:
    corpus_ = []
    stop_w = ['최고','하나','별로','있다', '없다', '진짜','좋다','카페' ,'이다', '좋다','아니다', '좀','같다','금방','보통', '조금', '굳', '뭐', '그', 
              '뭔', '항상', '알', '건', '더','이','내','두', '점','그냥','정말','그렇다','역시','안','수','거','만','것']

    for i in lda_train.preprocessed:
      sub1 = re.sub('라떼|라테|아메리카노', '커피', i)
      sub2 = re.sub('케이크|스콘|마카롱', '디저트', sub1)
      inner = []
      for t in sub2.split(' '):
        if t not in stop_w:
          inner.append(t)
      corpus_.append(' '.join(inner))
    
    corpus_gen = []
    for i in corpus_:
      inner = i.split(' ')
      corpus_gen.append(inner)

    return corpus_gen
```

```
make_corpus_for_gensim(lda_train.preprocessed)
```

```
from gensim import corpora
dictionary = corpora.Dictionary(corpus_gen)
dictionary.filter_extremes(no_below=8)
corpus = [dictionary.doc2bow(text) for text in corpus_gen]
print(corpus[1]) # 수행된 결과에서 두번째 뉴스 출력. 첫번째 문서의 인덱스는 0
```

```
import gensim
NUM_TOPICS = 4
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=450)
topics = ldamodel.print_topics(num_words=15)
for topic in topics:
    print(topic)
```
