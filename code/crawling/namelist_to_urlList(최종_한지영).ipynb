```
urllist = []
for i in namelist:
  url = 'https://search.naver.com/search.naver?where=blog&sm=tab_opt&query='+i+'&dup_remove=1&post_blogurl=&post_blogurl_without=&nso=so%3Add%2Ca%3At%2Cp%3Aall'
  url = url.replace("강남카페 ","강남 ")
  url = url.replace("건대카페 ","건대 ")
  url = url.replace("대학로카페 ","대학로 ")
  url = url.replace("홍대카페 ","홍대 ")
  url = url.replace(" ","%20")

  response = requests.get(url)
  html = response.text
  soup = BeautifulSoup(html, 'html.parser')
  my_titles = soup.select('a.api_txt_lines')
  a_tags = soup.find_all("a", {'class': "api_txt_lines total_tit"})

  for a_tag in a_tags[0:20]:
      # Tag의 href 속성 가져오기
      oneurl = a_tag.get('href')
      # 네이버 카페 거르기
      regex = re.compile("blog.naver.com/")
      mo = regex.search(oneurl)
      if mo != None:
        innerurllist = []
        innerurllist.append(i)
        innerurllist.append(oneurl)
        urllist.append(innerurllist)
  print(len(urllist))
```

```
import pandas as pd
df = pd.DataFrame(urllist)
df
```
